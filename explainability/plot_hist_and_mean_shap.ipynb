```python
import numpy as np
import csv
import os
import sys
import pandas as pd
from explainability_utils import plot_shap
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'..\preprocessed_dset_1\features_and_metadata.csv')

MAX_EVALS = 5000
shap_path = "shap_tensors"

tf64_1, tf32_1 = np.load(r"..\preprocessed_dset_1\sp_64\f_t_range.npy"), np.load(r"..\preprocessed_dset_1\sp_32\f_t_range.npy")
ft_1 = [tf64_1[:2], tf64_1[2:], tf32_1[:2], tf32_1[2:]]

tf64_2, tf32_2 = np.load(r"..\preprocessed_dset_2\sp_64\f_t_range.npy"), np.load(r"..\preprocessed_dset_2\sp_32\f_t_range.npy")
ft_2 = [tf64_2[:2], tf64_2[2:], tf32_2[:2], tf32_2[2:]]
```

```python
path = r"..\preprocessed_dset_1\sp_32\explainer_p_wave"
files = [f for f in os.listdir(os.path.join(path, "aftershock"))] + [f for f in os.listdir(os.path.join(path, "foreshock"))]

# Estrarre i nomi degli eventi dai file (senza .png e senza _pre/_post)
event_names = [f.replace(".png", "").replace("_pre", "").replace("_post", "") for f in files]

# Filtrare i dati per gli eventi presenti
filtered_df = df[df["trace_name"].isin(event_names)]

# Estrarre la colonna p_s_diff_sec e salvarla in un array NumPy
p_s_diff_sec_values_1 = filtered_df["p_s_diff_sec"].values + 5

# Plot del grafico a istogramma
plt.figure(figsize=(10, 6))
plt.hist(p_s_diff_sec_values_1, bins=100, edgecolor="black", alpha=0.7)
plt.xlabel("p_s_diff_sec")
plt.ylabel("Frequency")
plt.title("Histogram of p_s_diff_sec values")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```

```python
path = r"..\preprocessed_dset_2\sp_32\explainer_p_wave"
files = [f for f in os.listdir(os.path.join(path, "aftershock"))] + [f for f in os.listdir(os.path.join(path, "foreshock"))]

# Estrarre i nomi degli eventi dai file (senza .png e senza _pre/_post)
event_names = [f.replace(".png", "").replace("_pre", "").replace("_post", "") for f in files]

# Filtrare i dati per gli eventi presenti
filtered_df = df[df["trace_name"].isin(event_names)]

# Estrarre la colonna p_s_diff_sec e salvarla in un array NumPy
p_s_diff_sec_values_2 = filtered_df["p_s_diff_sec"].values + 5

# Plot del grafico a istogramma
plt.figure(figsize=(10, 6))
plt.hist(p_s_diff_sec_values_2, bins=100, edgecolor="black", alpha=0.7)
plt.xlabel("p_s_diff_sec")
plt.ylabel("Frequency")
plt.title("Histogram of p_s_diff_sec values")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```

## P-fixed models

### Slide 1: Model P64_08 (forshocks and aftershocks)

```python
ft = ft_1
pre = np.load(os.path.join(shap_path, f"foreshock_p64_08_{MAX_EVALS}.npy"))
post = np.load(os.path.join(shap_path, f"aftershock_p64_08_{MAX_EVALS}.npy"))
alpha_max = max(np.max(pre), np.max(post))
alpha_min = min(np.min(pre), np.min(post))
plot_shap(pre, onechannel=True, background=None, ft = ft, alpha_max=alpha_max, hist = p_s_diff_sec_values_1, alpha_min=alpha_min, spec_type = "p64_08", mean = f"Mean Shap Tensor of 300 correct predicted Foreshock Samples \n Model: P-wave fixed with 64_08 settings")
plot_shap(post, onechannel=True, background=None, ft = ft, alpha_max=alpha_max, hist = p_s_diff_sec_values_1, alpha_min=alpha_min, spec_type = "p64_08", mean = f"Mean Shap Tensor of 300 correct predicted Aftershock Samples \n Model: P-wave fixed with 64_08 settings")
```
